```python
import pandas as pd
import numpy as np
```

# **Latihan Binning **

```python
Data = pd.read_csv('Data_LatihanBinner - Sheet1.csv')
Df = pd.DataFrame(Data)
Df
```

1. Buatlah pengelompokan total transaksi
  Total Transaksi Kelompok Transaksi
  10.000 – 100.000 Biasa, 101.000 – 150.000 Menangah, 150.00 – 200.000 Menengah Atas, >= 200.000 Atas

```python
bins = [10000, 100000, 150000, 200000, 500000]
labels = ['Biasa', 'Menengah', 'Menengah Atas', 'Atas']
Df['Kelompok Transaksi'] = pd.cut(Df['Total'], bins=bins, labels=labels)
Df
```

2. Buatlah pengelompokan Jumlah pembelian
Jumlah Kelompok Jumlah
1 = J1, 2 = J2, 3 = J3, >= 4 J4

```python
bins = [0, 1, 2, 3, 4]
labels = ['J1', 'J2', 'J3', 'J4']
Df['Kelompok Jumlah'] = pd.cut(Df['Jumlah'], bins = bins, labels = labels)
Df
```

3. Buatlah kolom kategori Barang pada data diatas menggunakan case statement

```python
def kategoribarang(kb):
  if kb == 'Baju':
    return 'Pakaian'
  elif kb == 'Celana':
    return 'Pakaian'
  elif kb == 'Sepatu':
    return 'Aksesoris'
  elif kb == 'Tas':
    return 'Aksesoris'
  else:
    return 'Aksesoris'

Df['Kategori Barang'] = Df['Barang'].apply(kategoribarang)
Df
```
