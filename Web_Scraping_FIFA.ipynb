```python
data = pd.read_html('https://en.wikipedia.org/wiki/2022_FIFA_World_Cup')

dict_table = {}
for letter, i in zip(alphabet, range(9, 65, 7)):
  df = data[i]
  df.rename(columns = {df.columns[1]: 'Team'}, inplace = True)
  df.pop('Qualification')
  dict_table[f'Group {letter}'] = df
```

```python
dict_table['Group H']
```

```python
with open('dict_table', 'wb') as output:
  pickle.dump(dict_table, output)
```

```python
tables = [data[9], data[16], data[23], data[30], data[37], data[44], data[51], data[58]]
```

```python
combined_table = pd.concat(tables, ignore_index=True)
combined_table
combined_table_sort = combined_table.sort_values(by='Pts', ascending=False)
combined_table_sort
```
